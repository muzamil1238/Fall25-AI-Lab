# Data preprocessing and dealing with null values


```python
df = df.fillna('Unknown')
df.isnull().sum()
```

```python
df.dtypes
```

```python

df.apply(lambda x: pd.to_numeric(x, errors='coerce')).dtypes
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df = df.rename(columns={
    'brand': 'Brand',
    'perfume': 'Perfume',
    'type': 'Type',
    'category': 'Category',
    'target_audience': 'Target_Audience',
    'longevity': 'Longevity'
})

df.head()
```

```python
df = df.apply(lambda x: x.str.lower() if x.dtype == "object" else x)
df.head()
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
df.head()
```

# Changing data type

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
df = pd.read_csv("Perfumes_dataset.csv")
le = LabelEncoder()
for col in df.columns:
    df[col] = le.fit_transform(df[col])
df = df.astype(int)
df.dtypes
```
